```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter, FFMpegWriter

def update_fields(c, mu, dx, dt, M, kappa, boundary_condition):
    Nx, Ny = c.shape
    
    for i in range(1, Nx - 1):
        for j in range(1, Ny - 1):
            laplacian_c = (c[i + 1, j] + c[i - 1, j] + c[i, j + 1] + c[i, j - 1] - 4 * c[i, j]) / dx ** 2
            mu[i, j] = c[i, j] ** 3 - c[i, j] - kappa * laplacian_c

    # Boundary conditions
    if boundary_condition == 'Dirichlet':
        c[0, :] = 0.5
        c[-1, :] = 0.5
        c[:, 0] = 0.5
        c[:, -1] = 0.5
    elif boundary_condition == 'Neumann':
        c[0, :] = c[1, :]
        c[-1, :] = c[-2, :]
        c[:, 0] = c[:, 1]
        c[:, -1] = c[:, -2]
    elif boundary_condition == 'Periodic':
        c[0, :] = c[-1, :]
        c[-1, :] = c[0, :]
        c[:, 0] = c[:, -1]
        c[:, -1] = c[:, 0]
        
    for i in range(1, Nx - 1):
        for j in range(1, Ny - 1):
            laplacian_mu = (mu[i + 1, j] + mu[i - 1, j] + mu[i, j + 1] + mu[i, j - 1] - 4 * mu[i, j]) / dx ** 2
            c[i, j] += dt * M * laplacian_mu

# Parameters
Nx, Ny = 100, 100
dx = 1.0
dt = 0.01
M = 1.0
kappa = 0.02
Nt = 100
L = np.sqrt(Nx * Ny)

# Boundary condition
boundary_condition = 'Dirichlet'

# Initialize
rng = np.random.default_rng(12345)
noise = 0.1
c0 = 0.5
c = c0 + noise * rng.standard_normal((Nx, Ny))
mu = np.zeros((Nx, Ny))
c_list = [c.copy()]

# Time Evolution
for t in range(Nt):
    update_fields(c, mu, dx, dt, M, kappa, boundary_condition)
    c_list.append(c.copy())

# Relative Error Calculation
print('c = ', c.sum() * dx ** 2 / L ** 2)
print('relative_error = ', np.abs(c.sum() - c_list[0].sum()) / c_list[0].sum())

# Create and save animation
fig, ax = plt.subplots()
im = ax.imshow(c_list[0], cmap='RdBu_r', vmin=0.0, vmax=1.0)
cb = plt.colorbar(im)
cb.set_label(r'$c(x,y)$')
tx = ax.text(5, 45, f't=0, {boundary_condition}', color='white')

def animate(i):
    im.set_data(c_list[i])
    im.set_clim(0.0, 1.0)
    tx.set_text(f't={i * dt:.2f}, {boundary_condition}')
    return [im, tx]

ani = FuncAnimation(fig, animate, frames=Nt, interval=200, blit=True)

# Output as .gif and .avi
writer_gif = PillowWriter(fps=20)
writer_avi = FFMpegWriter(fps=20, metadata=dict(artist='Me'), bitrate=1800)
ani.save(f"bc-c0-{boundary_condition}_{c0}.gif", writer=writer_gif)
ani.save(f"bc-c0-{boundary_condition}_{c0}.avi", writer=writer_avi)

# Visualization of the final concentration
plt.imshow(c, cmap='RdBu_r', vmin=0.0, vmax=1.0)
plt.title(f'$c_0={c0:.1f}$')
plt.savefig(f'cahn-hilliard-c0-{c0:.1f}.png')
plt.show()
```
